```python
from pandas import read_csv
import seaborn as sns
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_regression,f_classif
dataframe=sns.load_dataset('tips')
dataframe
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dataframe['sex']=lb.fit_transform(dataframe['sex'])
dataframe['smoker']=lb.fit_transform(dataframe['smoker'])
dataframe['day']=lb.fit_transform(dataframe['day'])
dataframe['time']=lb.fit_transform(dataframe['time'])
```

```python
dataframe
```

```python
dataframe.info()
```

```python
#x is target column and y is independent one
X = dataframe.drop('tip',axis=1)
Y = dataframe.tip
```

```python
X
```

```python
Y
```

```python
#to select the best columns among all the columns we can use selectkbest method
test = SelectKBest(score_func=f_regression, k=3).fit(X,Y)
test
```

```python

#to get the statistics(scores) of our hypothesis tests
test.scores_
```

```python
#to get the round figure
np.round(test.scores_,3)
```

```python
dataframe.columns
```

```python
np.round(test.pvalues_,3)
```

```python
#to get the names of best columns
test.get_feature_names_out()
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['species']=lb.fit_transform(df['species'])
```

```python
df.info()
```

```python
X = df.drop('species',axis=1)
Y = df['species']
```

```python
X
```

```python
Y
```

```python
chi = SelectKBest(score_func=chi2, k=2).fit(X,Y)
chi
```

```python
chi.scores_
```

```python
np.round(chi.scores_,2)
```

```python
df.columns
```
